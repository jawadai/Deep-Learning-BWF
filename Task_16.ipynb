```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'id': np.random.randint(1, 6, 5), 'value1': np.random.randint(10, 51, 5)})

df2 = pd.DataFrame({'id': np.random.randint(1, 6, 5), 'value2': np.random.randint(100, 151, 5)})
```

```python
df1
```

```python
df2
```

```python
concatenated_df = pd.concat([df1, df2])
```

```python
concatenated_df
```

```python
merged_df = pd.merge(df1, df2, on = 'id')
```

```python
merged_df
```

```python
df = pd.DataFrame({'id': np.random.randint(1, 101, 1000),
                   'variable1': np.random.randint(1, 11, 1000),
                   'variable2': np.random.randint(11, 21, 1000),
                   'variable3': np.random.randint(21, 31, 1000),
                   'variable4': np.random.randint(31, 41, 1000)})
```

```python
df
```

```python
melted_df = pd.melt(df, id_vars=['id'], var_name='variable', value_name='value')

print(melted_df.head(10))
```

```python
balance_sheet = {
    'Particulars': ['Cash', 'Accounts Receivable', 'Inventory', 'Property, Plant and Equipment', 'Other Assets',
                    'Accounts Payable', 'Accrued Expenses', 'Long-Term Debt', 'Shareholders Equity'],
    '2021': [1000000, 500000, 2000000, 3000000, 500000, 1000000, 50000, 2000000, 1000000],
    '2022': [1200000, 600000, 2500000, 3500000, 700000, 1500000, 75000, 3000000, 1500000]
}

df = pd.DataFrame(balance_sheet)


```

```python
df
```

```python
df_melt = df.melt(id_vars='Particulars')
df_melt
```

```python
df_melt.pivot(index="Particulars",
              columns="Year",
              values="Amount").reset_index()

```
