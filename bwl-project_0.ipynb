```python
states_data_df['GenderPop'][20] = '2872643M_0F'
states_data_df['GenderPop'][26] = '510163M_0F' 
```

```python
states_data_df['Income'] = states_data_df['Income'].str.replace('$', '').astype(float)
```

```python
states_data_df[['MalePop', 'FemalePop']] = states_data_df['GenderPop'].str.split('_' , expand=True)
```

```python
states_data_df
```

```python
states_data_df.reset_index(drop=True, inplace=True)
```

```python
states_data_df
```

```python
states_data_df.loc[20, 'FemalePop'] = "000F"
states_data_df.loc[26, 'FemalePop'] = "000F"
```

```python
states_data_df['MalePop'] = states_data_df['MalePop'].str.replace('M','').astype(int)
states_data_df['FemalePop'] = states_data_df['FemalePop'].str.replace('F','').astype(int)
```

```python
states_data_df
```

```python
states_data_df.head(2)
```

```python
def changeIntoNumeric(colName, sign):
    states_data_df[colName] = states_data_df[colName].str.replace(sign, '').astype(float)
```

```python
states_data_df
```

```python
changeIntoNumeric('Hispanic', '%')
changeIntoNumeric('White', '%')
changeIntoNumeric('Black', '%')
changeIntoNumeric('Native', '%')
changeIntoNumeric('Asian', '%')
changeIntoNumeric('Pacific', '%')
```
